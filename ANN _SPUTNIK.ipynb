##### Function Definition

Here we need to define some functions that will be used in building our neural network.

```python
# define a quadratic function
def quad(x,a,b,c):
    return a*x**2 + b*x + c

# define a cubic function
def cubic(x,a,b,c,d):
    return a*x**3 + b*x**2 + c*x + d

# define a exponential function
def exp(x,a):
    return a**x

def linear(x,m,c):
    return m*x + c
```

```python
import tensorflow 
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import load_img,img_to_array
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K
```

```python
data_gen = ImageDataGenerator(rotation_range=40,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             rescale=1./255,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             fill_mode='nearest')
```

```python
# define a function to generate the n plots each for training set
def gen(name,prefix,location,n):
    pix = load_img(name)
    pix_to_array = img_to_array(pix)

    pix_to_array = pix_to_array.reshape((1,) + pix_to_array.shape)
    pix_to_array.shape

# generate 600 images for the sine function
    count = 0
    for i in data_gen.flow(pix_to_array, batch_size=1,save_to_dir= location,
                      save_prefix=prefix, save_format='jpg'):
        count +=1
        if count == n:
            break
```

```python
# generate images for the exponential
gen(name="Desktop/Sputnik/imagefiles/exponential.jpg",prefix="exponent", 
    location="Desktop/Sputnik/generated images/exponent", n = 600)

#generate images for the cubic function
gen(name="Desktop/Sputnik/imagefiles/cubic.jpg",prefix="cubic",
    location="Desktop/Sputnik/generated images/cubic", n = 600)

# generate for cosine
gen(name="Desktop/Sputnik/imagefiles/cosine.jpg",prefix="cosine",
    location="Desktop/Sputnik/generated images/cosine", n = 600)

# generate for the sine
gen(name="Desktop/Sputnik/imagefiles/sine.jpg",prefix="sine",
    location="Desktop/Sputnik/generated images/sine", n = 600)

# generate for the quadratic
gen(name="Desktop/Sputnik/imagefiles/quadratic.jpg",prefix="quad",
    location="Desktop/Sputnik/generated images/quadratic", n = 600)

# generate for the linear
gen(name="Desktop/Sputnik/imagefiles/linear.jpg",prefix="linear",
    location="Desktop/Sputnik/generated images/linear", n = 600)



```

```python
# generate for the linear
gen(name="Desktop/Sputnik/imagefiles/linear.jpg",prefix="linear",
    location="Desktop/Sputnik/validation/linear",n=180)

# generate images for the exponential
gen(name="Desktop/Sputnik/imagefiles/exponential.jpg",prefix="exponent", 
    location="Desktop/Sputnik/validation/exponent", n = 180)

#generate images for the cubic function
gen(name="Desktop/Sputnik/imagefiles/cubic.jpg",prefix="cubic",
    location="Desktop/Sputnik/validation/cubic", n = 180)

# generate for cosine
gen(name="Desktop/Sputnik/imagefiles/cosine.jpg",prefix="cosine",
    location="Desktop/Sputnik/validation/cosine", n = 180)

# generate for the sine
gen(name="Desktop/Sputnik/imagefiles/sine.jpg",prefix="sine",
    location="Desktop/Sputnik/validation/sine", n = 180)

# generate for the quadratic
gen(name="Desktop/Sputnik/imagefiles/quadratic.jpg",prefix="quad",
    location="Desktop/Sputnik/validation/quadratic", n = 180)
```

```python
# dimensions of our images.
img_width, img_height = 150, 150

train_data_dir = "Desktop/Sputnik/generated images"
validation_data_dir = "Desktop/Sputnik/validation"
nb_train_samples = 3000
nb_validation_samples = 1000
epochs = 50
batch_size = 16
```

```python

if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
```

### Build the architecture

```python

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

### Compile 

```python

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
# this is the augmentation configuration we will use for testing:
# only rescaling
test_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = data_gen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')
```

```python
validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')
```

```python
model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size)
```
